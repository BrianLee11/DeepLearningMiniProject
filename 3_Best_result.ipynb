```python
import numpy as np

def compute_channel_stats(data):
    """
    Compute the mean and standard deviation for each color channel in the image data.

    Parameters:
    data (numpy.ndarray): A 2D numpy array where each row represents an image (num_images, channel*height*width).

    Returns:
    dict: A dictionary with means and standard deviations for each channel.
    """
    if data.ndim != 2:
        raise ValueError("Input data must be a 2D array.")

    # Reshape data to (num_images, channels, height, width)
    reshaped_data = data.reshape(data.shape[0], 3, 32, 32)

    # Compute means and stds
    means = reshaped_data.mean(axis=(0, 2, 3))
    stds = reshaped_data.std(axis=(0, 2, 3))

    return {'means': means, 'stds': stds}

# Assuming train_data_max is a numpy array with shape (60000, 3072)
# Example usage:
stats = compute_channel_stats(train_data_max)
print("Channel means:", stats['means'])
print("Channel standard deviations:", stats['stds'])
```

```python
def normalize_per_channel(data):
    """
    Normalize each channel of the image data.

    Parameters:
    data (numpy.ndarray): A 2D numpy array where each row represents an image (num_images, channel*height*width).

    Returns:
    numpy.ndarray: The normalized data.
    """
    if data.ndim != 2:
        raise ValueError("Input data must be a 2D array.")

    # Initialize the array to store normalized data
    normalized_data = np.zeros_like(data, dtype=np.float32)

    # Calculate mean and std for each channel and normalize
    for i in range(3):  # Assuming 3 channels
        channel_data = data[:, i * 1024:(i + 1) * 1024]
        mean = channel_data.mean(axis=1).reshape(-1, 1)
        std = channel_data.std(axis=1).reshape(-1, 1)
        normalized_data[:, i * 1024:(i + 1) * 1024] = (channel_data - mean) / (std + 1e-8)  # Adding a small epsilon to avoid division by zero

    return normalized_data
```

Train, Test function implementation

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def train_resnetPreset(batch_combined_data, batch_combined_labels):
    # Reshape and normalize the data
    batch_combined_data = batch_combined_data.reshape((-1, 32, 32, 3)) / 255.0

    # Convert labels to categorical
    batch_combined_labels = tf.keras.utils.to_categorical(batch_combined_labels, 10)

    # Define the ResNet model
    model = tf.keras.applications.ResNet50(
        include_top=False,
        weights=None,
        input_shape=(32, 32, 3),
        classes=10,
    )

    # Adding the final layers to the model
    model = models.Sequential([
        model,
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dense(10, activation='softmax'),
    ])

    # Compile the model
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy'],
    )

    # Train the model
    model.fit(
        batch_combined_data,
        batch_combined_labels,
        epochs=10,  # You can adjust the number of epochs
        batch_size=64,  # And the batch size
        validation_split=0.2,  # Using 20% of data for validation
    )

    return model

# Call the function with your data
# model = train_resnet(batch_combined_data, batch_combined_labels)
```

```python
trained_model_resnetPreset = train_resnetPreset(train_data, train_labels)
```

```python
def test_resnet(model, test_data, test_labels):
    # Reshape and normalize the test data
    test_data = test_data.reshape((-1, 32, 32, 3)) / 255.0

    # Convert test labels to categorical
    test_labels = tf.keras.utils.to_categorical(test_labels, 10)

    # Evaluate the model on the test data
    test_loss, test_accuracy = model.evaluate(test_data, test_labels)

    print(f"Test Loss: {test_loss}")
    print(f"Test Accuracy: {test_accuracy}")

    return test_accuracy
```

Model function ResidualConvet

```python
import torch.optim as optim  # Make sure this import is included
from torch.utils.data import TensorDataset, DataLoader
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ResidualConvnet(nn.Module):
    def __init__(self, input_channels=1, num_classes=10):
        super(ResidualConvnet, self).__init__()

        # First layer
        self.first_layer = nn.Sequential(
            nn.Conv2d(in_channels=input_channels, out_channels=64, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.BatchNorm2d(64)
        )

        # First residual block
        self.first_block = nn.Sequential(
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(64),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(64)
        )
        self.first_BatchNorm_MaxPool = nn.Sequential(
            nn.BatchNorm2d(64),
            nn.MaxPool2d(2, 2)
        )

        # Second residual block
        self.second_block = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(128),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(128)
        )
        self.second_BatchNorm_MaxPool = nn.Sequential(
            nn.BatchNorm2d(128),
            nn.MaxPool2d(2, 2)
        )
        self.conv_64_128 = nn.Conv2d(64, 128, kernel_size=1)  # Adjusted for matching dimensions

        # Third residual block
        self.third_block = nn.Sequential(
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(256),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(256)
        )
        self.thrid_BatchNorm_MaxPool = nn.Sequential(
            nn.BatchNorm2d(256),
            nn.MaxPool2d(2, 2)
        )
        self.conv_128_256 = nn.Conv2d(128, 256, kernel_size=1)  # Adjusted for matching dimensions

        # Adaptive Pooling and Projection
        self.adaptive_pool = nn.AdaptiveAvgPool2d((1, 1))
        self.projection = nn.Linear(256, num_classes)

    def forward(self, x):
        # Initial layer
        x = self.first_layer(x)

        # First block with residual connection
        identity = x
        x = self.first_block(x) + identity
        x = self.first_BatchNorm_MaxPool(x)

        # Second block with residual connection
        identity = self.conv_64_128(x)  # Adjust identity to match dimensions
        x = self.second_block(x) + identity
        x = self.second_BatchNorm_MaxPool(x)

        # Third block with residual connection
        identity = self.conv_128_256(x)  # Adjust identity to match dimensions
        x = self.third_block(x) + identity
        x = self.thrid_BatchNorm_MaxPool(x)

        # Adaptive Pooling and Projection
        x = self.adaptive_pool(x)
        x = torch.flatten(x, 1)
        x = self.projection(x)

        return x

    def train_model(self, batch_combined_data, batch_combined_labels, num_epochs=10, batch_size=32, learning_rate=0.001, step_size=1, gamma=0.1):
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.to(device)

        # Assuming the data is in the format [batch_size, 3072] for 3x32x32 images
        # Convert numpy arrays to tensors and create a DataLoader
        X = torch.tensor(batch_combined_data.reshape(-1, 3, 32, 32), dtype=torch.float32)
        y = torch.tensor(batch_combined_labels, dtype=torch.long)

        dataset = TensorDataset(X, y)
        dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

        # Define a loss function and optimizer
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(self.parameters(), lr=learning_rate)
        scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=step_size, gamma=gamma)

        # Clearing CUDA cache
        torch.cuda.empty_cache()

        # Training loop
        for epoch in range(num_epochs):
            for inputs, labels in dataloader:
                inputs, labels = inputs.to(device), labels.to(device)

                # Normalize data inside the loop to avoid memory overhead
                inputs /= 255.0

                optimizer.zero_grad()
                outputs = self(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()

            scheduler.step()
            print(f'Epoch {epoch + 1}/{num_epochs}, Learning Rate: {scheduler.get_last_lr()}, Loss: {loss.item()}')

        return self
```

```python
def count_parameters(model):
    """
    Counts the number of trainable parameters in a PyTorch model.

    Args:
        model (nn.Module): The PyTorch model.

    Returns:
        int: Total number of trainable parameters in the model.
    """
    return sum(param.numel() for param in model.parameters() if param.requires_grad)
```

```python
def test_resnet_ResidualCovnet(model, test_data, test_labels, batch_size=32):
    model.eval()  # Set the model to evaluation mode
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # Convert test data and labels to tensors
    test_data = torch.tensor(test_data.reshape(-1, 3, 32, 32), dtype=torch.float32).to(device) / 255.0
    test_labels = torch.tensor(test_labels, dtype=torch.long).to(device)

    correct = 0
    total = 0

    with torch.no_grad():
        for i in range(0, len(test_data), batch_size):
            # Process the test data in batches
            batch_data = test_data[i:i+batch_size]
            batch_labels = test_labels[i:i+batch_size]

            outputs = model(batch_data)
            _, predicted = torch.max(outputs.data, 1)
            total += batch_labels.size(0)
            correct += (predicted == batch_labels).sum().item()

    accuracy = correct / total
    return accuracy
```

```python
def predict_resnet(model, predict_data):
    # Reshape and normalize the test data
    predict_data = predict_data.reshape((-1, 32, 32, 3)) / 255.0

    # Predict the labels
    predictions = model.predict(predict_data)

    # Convert predicted probabilities to class labels
    predict_labels = np.argmax(predictions, axis=1)

    return predict_labels
```

```python
def predict_resnet_ResidualCovnet(model, predict_data, batch_size=32):
    model.eval()  # Set the model to evaluation mode
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # Initialize an empty list to store the predictions
    predictions = []

    # Split the data into batches
    for i in range(0, len(predict_data), batch_size):
        # Select the data for the current batch
        batch_data = predict_data[i:i+batch_size]

        # Convert to tensor and reshape
        batch_data = batch_data.reshape(-1, 3, 32, 32)
        batch_data = torch.tensor(batch_data, dtype=torch.float32).to(device) / 255.0

        # Perform the prediction
        with torch.no_grad():
            outputs = model(batch_data)
            _, batch_predictions = torch.max(outputs, 1)

        # Move the batch predictions to CPU and convert to numpy
        batch_predictions = batch_predictions.cpu().numpy()

        # Store the predictions
        predictions.extend(batch_predictions)

    return np.array(predictions)
```

Normalization

```python
import torch
import numpy as np

def predict_resnet_ResidualCovnet(model, predict_data, batch_size=32):
    model.eval()  # Set the model to evaluation mode
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # Mean and std deviation for normalization, adjust these values as per your training data
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])

    # Initialize an empty list to store the predictions
    predictions = []

    # Split the data into batches
    for i in range(0, len(predict_data), batch_size):
        # Select the data for the current batch
        batch_data = predict_data[i:i + batch_size]

        # Convert to tensor and reshape
        batch_data = batch_data.reshape(-1, 3, 32, 32)
        batch_data = torch.tensor(batch_data, dtype=torch.float32).to(device)

        # Normalize the data
        for c in range(3):
            batch_data[:, c, :, :] = (batch_data[:, c, :, :] / 255.0 - mean[c]) / std[c]

        # Perform the prediction
        with torch.no_grad():
            outputs = model(batch_data)
            _, batch_predictions = torch.max(outputs, 1)

        # Move the batch predictions to CPU and convert to numpy
        batch_predictions = batch_predictions.cpu().numpy()

        # Store the predictions
        predictions.extend(batch_predictions)

    return np.array(predictions)
```
